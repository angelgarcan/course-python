```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
type(titanic)
titanic
```

```python
iris = sns.load_dataset("iris")
type(iris)
iris
```

```python
sns.scatterplot(x=iris.petal_length,y=iris.petal_width)
```

```python
sns.lineplot(x=iris.petal_length,y=iris.petal_width)
```

```python
sns.stripplot(x="species", y="petal_length", data=iris)
```

```python
sns.swarmplot(x="species", y="petal_length", data=iris)
```

```python
sns.boxplot(x="alive", y="age", hue="adult_male", data=titanic)
```

```python
sns.countplot(x="deck", data=titanic, palette="Greens_d")
```

```python
sns.pointplot(x="class", y="survived", hue="sex", data=titanic, palette={"male": "g", "female": "m"},
              markers=["^", "o"], linestyles=["-", "--"])
```

```python
sns.barplot(x="sex", y="survived", hue="class", data=titanic)
```
