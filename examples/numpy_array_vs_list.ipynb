```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bts=[]
ats=[]

x=np.arange(1,1000,10)
for n in x:
    bt=%timeit -q -o -r200 -n5 list(map(lambda x:x*2,range(n)))
    bts.append(bt.average)
    at=%timeit -q -o -r200 -n5 np.arange(n)*2
    ats.append(at.average)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(x, bts, label = "List time")
plt.plot(x, ats, label = "Array time")
plt.xlabel('n elements')
plt.ylabel('time')
plt.legend()
plt.show()
```

```python
i=30
plt.plot(x[:i], bts[:i], label = "List time", marker = 'o')
plt.plot(x[:i], ats[:i], label = "Array time", marker = 'o')
plt.xlabel('n elements')
plt.ylabel('time')
plt.legend()
plt.show()
```
